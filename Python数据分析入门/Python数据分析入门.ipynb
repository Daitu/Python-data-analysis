```python
## 定义一个计算1～x的累加和的函数
def sumx(x):
    xsum = 0
    if x < 1:
        print("请输入一个更大的数")
        xsum = None
    else:
        x = range(1,x+1)  ## 生成1～x的向量
        for ii in x:
            xsum = xsum+ii
    return xsum
```

#### Python数据探索与预处理

主要是处理数据的缺失值

```python
## 读取带有缺失值的数据
Irisdf = pd.read_csv("data/Iris带缺失值.csv")
Irisdf.head()
```

```python
## 查看数据信息
Irisdf.info()
```

```python
## 查看数据的描述统计
Irisdf.describe()
```

```python
## 使用可视化的方法观察数据集的缺失值情况
import missingno as msno 
msno.matrix(Irisdf,figsize=(14, 7),width_ratios=(10, 4),color=(0.25, 0.25, 0.5))
plt.show()
```

```python
## 缺失值填充的相关方式，1:删除带有缺失值的样本
Irisdf2 = Irisdf.dropna(axis=0)
msno.matrix(Irisdf2,figsize=(14, 7),
            width_ratios=(10, 4),color=(0.25, 0.25, 0.5))
plt.show()
## 数据只剩下121个样本，不存在缺失值
```

```python
## 缺失值填充的相关方式，2:使用均值填充
## 为了方便观察数据填充效果，使用散点图对数据进行可视化

## 找到缺失值所在的位置（针对SepalLengthCm变量和SepalWidthCm变量）
nanaindex = pd.isna(Irisdf.SepalLengthCm) | pd.isna(Irisdf.SepalWidthCm)
## 使用变量均值进行填充,计算均值
SepalLmean = Irisdf.SepalLengthCm.mean()
SepalWmean = Irisdf.SepalWidthCm.mean()
## 填充
SepalLength = Irisdf.SepalLengthCm.fillna(value = SepalLmean)
SepalWidth = Irisdf.SepalWidthCm.fillna(value = SepalWmean)
```

```python
## 可视化填充后的结果
plt.figure(figsize = (14,7))
plt.subplot(1,2,1)   # 可视化原始数据
plt.scatter(Iris.SepalLengthCm[~nanaindex],Iris.SepalWidthCm[~nanaindex],
            c = "blue",marker = "o",label = "非缺失值")
plt.scatter(Iris.SepalLengthCm[nanaindex],Iris.SepalWidthCm[nanaindex],
            c = "red",marker = "s",label = "待填充数据")
plt.grid()
plt.legend(loc = "upper right",fontsize = 12)
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("原始数据")

plt.subplot(1,2,2)   # 可视化填充了缺失值的数据
## 不带缺失值的样本散点图
plt.scatter(SepalLength[~nanaindex],SepalWidth[~nanaindex],
            c = "blue",marker = "o",label = "非缺失值")
## 带缺失值的样本散点图
plt.scatter(SepalLength[nanaindex],SepalWidth[nanaindex],
            c = "red",marker = "s",label = "填充的缺失值")
plt.grid()
plt.legend(loc = "upper right",fontsize = 12)
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("使用变量均值填充")

plt.tight_layout()
plt.show()

## 可以发现待填充待数据分布上不均匀，而均值填充的结果在两条直线上，数据填充效果并不是很好
```

```python
## 缺失值填充的相关方式，3:使用K近邻的方式进数据填充，可以综合数据中的多个变量的情况
## 为了方便观察数据填充效果，使用散点图对数据进行可视化
from sklearn.impute import KNNImputer
## KNNImputer缺失值填补方法
knnimp = KNNImputer(n_neighbors = 6)
Irisdfknn = knnimp.fit_transform(Irisdf.drop(["Id","Species"],axis = 1))
## 获取填充后的变量
SepalLength = Irisdfknn[:,0]
SepalWidth = Irisdfknn[:,1]
```

```python
## 可视化填充后的结果
plt.figure(figsize = (14,7))
plt.subplot(1,2,1)   # 可视化原始数据
plt.scatter(Iris.SepalLengthCm[~nanaindex],Iris.SepalWidthCm[~nanaindex],
            c = "blue",marker = "o",label = "非缺失值")
plt.scatter(Iris.SepalLengthCm[nanaindex],Iris.SepalWidthCm[nanaindex],
            c = "red",marker = "s",label = "待填充数据")
plt.grid()
plt.legend(loc = "upper right",fontsize = 12)
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("原始数据")

plt.subplot(1,2,2)   # 可视化填充了缺失值的数据
## 不带缺失值的样本散点图
plt.scatter(SepalLength[~nanaindex],SepalWidth[~nanaindex],
            c = "blue",marker = "o",label = "非缺失值")
## 带缺失值的样本散点图
plt.scatter(SepalLength[nanaindex],SepalWidth[nanaindex],
            c = "red",marker = "s",label = "填充的缺失值")
plt.grid()
plt.legend(loc = "upper right",fontsize = 12)
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("使用KNN填充")

plt.tight_layout()
plt.show()

## 而KNN填充的结果和待填充的缺失值数据很接近
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
## 建立一元线性回归模型
lm1 = smf.ols("PetalWidthCm ~ PetalLengthCm",data=Iris2).fit()
print(lm1.summary())
```

**R^2 : 0.927，很接近于1，说明线性回归拟合效果较好**

**Prob (F-statistic): 5.78e-86 ； 说明模型是显著的**

**回归系数t检验P值，远小于0.05，说明回归系数是显著的**

**回归模型为:**

$PetalWidthCm = 0.4164 * PetalLengthCm -0.3665$

```python
## 拓展思考内容:对数据进行回归模型中加入分类变量，效果会怎么样？
plt.figure(figsize=(10,6))
sns.lmplot(data = Iris2,x = "PetalLengthCm",y = "PetalWidthCm",hue="Species",
           height = 6,aspect=1.7)

## 把一元回归的结果添加到图像中进行对比
X = pd.DataFrame(data = np.arange(0.8,7.2,step = 0.1),
                 columns= ["PetalLengthCm"])
Y = lm1.predict(X)  # 使用一元回归模型获取预测值
plt.plot(X,Y,"r-",linewidth = 3)

plt.grid()
plt.show()

## 直观的感觉两种回归结果的差异性还是很大的
```

#####  使用petal的两个属性利用LDA分类器建立分类模型


```python
## 数据准备
X = Iris2.iloc[:,0:4].values
y = Iris2.Species.values
## 将类别标签的字符串转化为
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
yle = LE.fit_transform(y)
print(X.shape)
print(y[0:5],y[50:55])
print(yle[0:5],yle[50:55])
```

```python
## 数据切分为训练集和测试集
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, yle, test_size=0.33, random_state=124)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
##训练LDA分类器
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
lda = LinearDiscriminantAnalysis()
## 训练模型
lda = lda.fit(X_train[:,2:4],y_train)
## 预测测试数据
y_pre = lda.predict(X_test[:,2:4])
print("使用两个特征在测试集上的预测精度为:",accuracy_score(y_test,y_pre))
```

```python
## 在二维空间中可视化测试集的分界面
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(10,6))
plot_decision_regions(X_test[:,2:4],y_test,lda,legend=2)
plt.show()
```

```python
## 使用4个特征训练LDA分类器，查看效果是否比使用2个特征的效果好
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score
lda = LinearDiscriminantAnalysis()
## 训练模型
lda = lda.fit(X_train,y_train)
## 预测测试数据
y_pre = lda.predict(X_test)
print("使用4个特征在测试集上的预测精度为:",accuracy_score(y_test,y_pre))

## 可以发现如果使用4个特征模型的预测精度更高
```
